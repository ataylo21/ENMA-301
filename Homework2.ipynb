```python
K = 1.38E-23
print('K=', K)
t_range = [100, 200, 300, 400, 600, 1000]
print('t-range:', t_range)
a = np.linspace(-0.3, 0.3, 50)
print('Possible energy values:', a)
b = np.empty_like(a)
print('Empty Array:', b)
```

Question 1 part 2

```python
def fE(energy_diff, t_range):
    k_B = 8.6173303e-5  # Boltzmann constant in eV/K
    return 1 / (1 + np.exp(energy_diff / (k_B * t_range)))
```

Question 1 part 3

```python
t_range = [100, 200, 300, 400, 600, 1000]
energy_diff = a-b
for T in t_range:
    if T<=100:
        one = fE(energy_diff, 100)
        print(one)
    elif T<=200:
        two = fE(energy_diff, 200)
        print(two)
    elif T<=300:
        three = fE(energy_diff, 300)
        print(three)
    elif T<=400:
        four = fE(energy_diff, 400) 
        print(four)
    elif T<=600:
        six = fE(energy_diff, 600)
        print(six)
    else:
        ten = fE(energy_diff, 1000)
        print(ten)     
```

```python
plt.xlabel('Fermi Energy (eV/K)')
plt.ylabel('Electron Distribution')
plt.title('Fermi Energy vs. Temperature')
plt.plot(one, label = 'T=100K')
plt.plot(two, label = 'T=200K')
plt.plot(three, label = 'T=300K')
plt.plot(four, label = 'T=400K')
plt.plot(six, label = 'T=600K')
plt.plot(ten, label = 'T=1000K')
plt.legend()
```

How would the fermi distribution look if the temperature was 0K? What would be the energy of the highest energy electron?

When the temperature is at 0K, the fermi distribution will be undifined. If you were to plug zero in for the temperature, you would be dividin by zero in the fermi energy function above. This means that at 0K, the electrons will be in lower energy states. The energy of the highest energy electron would be the fermi-level, and it would be located at where the electron distribution is at 0K which is at the 1, since no electrons seem to be occupying energy states higher than that.

Question 2 part 2

```python
#for intrinsic semiconductors n = p = ni
#Conductivity = ni|e|(electron mobility + hole mobility)
def intrinsic_charge_carriers(conductivity, electron_mobility, hole_mobility):
    # Constants
    q = 1.6e-19  # Elementary charge (Coulombs)
    ni = conductivity / (q * (electron_mobility + hole_mobility))
    return ni
```

```python
#calculations for getting ni values
#mat_labels = ['Si','Ge','GaAs','InSb']
#bandgap in order of mat_label = (1.11, 0.67, 1.42, 0.17)
ni_Si = intrinsic_charge_carriers(Si[1], Si[2], Si[3])
ni_Ge = intrinsic_charge_carriers(Ge[1], Ge[2], Ge[3])
ni_GaAs = intrinsic_charge_carriers(GaAs[1], GaAs[2], GaAs[3])
ni_InSb = intrinsic_charge_carriers(InSb[1], InSb[2], InSb[3])
ni_values = (ni_Si, ni_Ge, ni_GaAs, ni_InSb)
print(ni_values)
bandgap = (1.11, 0.67, 1.42, 0.17)
print(bandgap)
```

```python
k = 8.62E-05
dopant = 1E16
def N(Nc, Nv, temperature): # Nc and Nv @ 300K
    Nc = Nc * (temperature / 300) ** (3/2)
    Nv = Nv * (temperature / 300) ** (3/2)
    return Nc, Nv
def ni(Nc, Nv, temperature, band): # N = Nc *Nv
    ni = (Nc * Nv) ** (1/2) * (2.71828  ** (-band / (2 * k * temperature))) #2.71828 = e^#
    return ni
```

Question 3 part 5 - If these extrinsic semiconductors are to be utilized in a high temperature application, at what temperature would you run into problems (from intrinsic carriers) for all 3 materials. Which material would be best for this high temp application? Explain briefly how this relates to the bandgap of each material.

Since these extrinsic semiconductors are to be used at high temperatures, at 800K I would start to run into problems for Si, 300K for Ge, and 1500K for GaAs. I think Si would be the best material for this high temperature application. Since Ge has a very low bandgap, it would be easier for electrons to jump when excited at the higher temperatures, which would be good for this application, except that a lot of the electrons would then be trying to jump at once, making it not as suitable for this appliction in comparison to Si. GaAs on the other hand, has a greater bandgap, but then the electrons would require more energy to be excited enough to go over the bandgap. Si has a good bandgap in between, so it will still be fairly easy to excite the electrons and have a good distance in bandgap for the electrons to pass so there won't be as much traffic.
